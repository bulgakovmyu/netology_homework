```python
text = ''
with open("D:\\GitRepo\\netology_homework\\data\\WoW\\tom_1.txt", 'r', encoding='utf-8') as f:
    buf = f.read().lower()
buf = re.sub(r'[^а-я ]','',buf)
text += re.sub(r' +',' ',buf)

with open("D:\\GitRepo\\netology_homework\\data\\WoW\\tom_2.txt", 'r', encoding='utf-8') as f:
    buf = f.read().lower()
buf = re.sub(r'[^а-я ]','',buf)
text += re.sub(r' +',' ',buf)

with open("D:\\GitRepo\\netology_homework\\data\\WoW\\tom_3.txt", 'r', encoding='utf-8') as f:
    buf = f.read().lower()
buf = re.sub(r'[^а-я ]','',buf)
text += re.sub(r' +',' ',buf)

with open("D:\\GitRepo\\netology_homework\\data\\WoW\\tom_4.txt", 'r', encoding='utf-8') as f:
    buf = f.read().lower()
buf = re.sub(r'[^а-я ]','',buf)
text += re.sub(r' +',' ',buf)

len(text)
```

```python
RUSSIAN = 'абвгдежзийклмнопрстуфхцчшщъыьэюя '
POS = {l: i for i, l in enumerate(RUSSIAN)}
len(POS)
```

```python
probabilities = np.zeros((len(RUSSIAN), len(RUSSIAN)))

for cp, cn in zip(text[:-1], text[1:]):
    probabilities[POS[cp], POS[cn]] += 1
    
probabilities = (probabilities / probabilities.sum(axis=1)[:, None])
```

```python
df = pd.DataFrame(probabilities, index=(c for c in RUSSIAN), columns=(c for c in RUSSIAN))
df
```

Вероятность слова из четырех букв, начинающегося  с "П" и заканчивающегося на "Р":

```python
start = POS['п']
end = POS['р']
prior = 1. / (len(RUSSIAN)-1) # без учета пробела
total_prob = 0
for c1 in RUSSIAN[:-1]:
    for c2 in RUSSIAN[:-1]:
        proba = prior * prior * probabilities[start, POS[c1]] * probabilities[POS[c1], POS[c2]] * probabilities[POS[c2], end]
        if proba > 0:
            print(''.join(['п', c1, c2, 'р']), proba)
            total_prob += proba
print(total_prob)
```

Слово, начинающееся с буквы “Б” и имеющее длину 8-мь символов:

```python
def word_gen(start,length):
    s = []
    s.append(start)
    i = 0
    while (i < length-1):
        index = POS[s[i]]
        next_char = np.random.choice(list(RUSSIAN), p=probabilities[index])
        if (next_char != ' '):
            s.append(next_char)
            i += 1

    return ''.join(s)
```

2. Прогноз с НММ

```python
from hmmlearn.hmm import GaussianHMM
```

```python
series = pd.read_csv("D:\\GitRepo\\Нетология\\Временные ряды\\1._Лекция\\Series\\mean-monthly-air-temperature-deg.csv")["Deg"]
```

```python
values = series.values.reshape(-1,1)
```

Выберем n с наименьшим score:

```python
best_n = 1
best_score = np.inf
for n in range(1, 14):
    model = GaussianHMM(n_components=n,
                        covariance_type="diag",
                        n_iter=1000)
    model.fit(values)
    if (model.score(values) < best_score):
        best_n = n
        best_model = model

print (best_n, best_model.score(values))
```

Построим проогноз на модели с 13 скрытыми компонентами:

```python
model = GaussianHMM(n_components=13,
                        covariance_type="diag",
                        n_iter=10000)
model.fit(values)
labels = model.predict(values)
means = np.zeros_like(values)
for i in range(model.n_components):
    means[labels == i] = model.means_[i]
    
plt.figure(figsize=(12, 6))
plt.plot(values)
plt.plot(means, linewidth=3)
```
